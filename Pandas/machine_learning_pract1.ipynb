Missing Values difference- isnull(), notnull(), dropna(), fillna(),  replace(), interpolate()


```python
df = pd.DataFrame({'A': [1, 2, None, 4]})
print(df['A'].isnull())
```

```python
df.dropna(inplace=True)
df.head()
```

```python
df['A'].fillna(0, inplace=True)
df.head()
```

```python
df.replace({1: 'one', 2: 'two'})
df.head()
```

```python
df['A'].interpolate(method='linear', inplace=True)
df.head()
```
